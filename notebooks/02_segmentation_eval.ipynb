## Inferencia sobre imagenes de test

```python
test_images = list(Path('../data/raw/demo').glob('*.jpg'))[:10]
print(f'Test images: {len(test_images)}')
```

```python
def visualize_segmentation(image_path):
    img = Image.open(image_path).convert('RGB')
    seg_map = segmenter.predict(img)
    mask = segmenter.get_garment_mask(seg_map)
    mask = clean_mask(mask)
    mask = largest_connected_component(mask)
    segmented = crop_to_content(apply_mask(img, mask))
    
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    axes[0].imshow(img); axes[0].set_title('Original'); axes[0].axis('off')
    axes[1].imshow(mask, cmap='gray'); axes[1].set_title('Mascara'); axes[1].axis('off')
    axes[2].imshow(segmented); axes[2].set_title('Segmentada'); axes[2].axis('off')
    plt.tight_layout(); plt.show()

for p in test_images[:5]:
    visualize_segmentation(p)
```
